<h1> Mempelajari Bar Plot / Bar Chart

```python
# Import Modules
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

```python
data = [25, 50, 80, 125, 200]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

<h3>Pengaturan Grid dan Color

```python
data = [25, 50, 80, 125, 200]
kategori = ['A', 'B', 'C', 'D', 'E']

plt.bar(kategori,
        data,
        color='green',
        alpha=0.50)

plt.grid(linestyle='--',
         linewidth=3,
         axis='y', # pilihan: x, y, both
         alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

<h3>Horizontal Bar Plot

```python
data = [25, 50, 80, 125, 200]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
# 2 pilihan 
# plt.barh(kategori, data) # 1. dimulai dari parameter terakhir 
plt.barh(kategori[::-1], data[::-1]) # 2. dimulai dari parameter pertama

plt.grid(linestyle='--',
         linewidth=1,
         axis='x',
         alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('kategori')
plt.title('Contoh Horizontal Bar Plot')

plt.show()
```

<h3>Stacked Bar Plot

```python
data1 = [40, 10, 75, 35, 100]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
# 2 Pilihan
plt.bar(kategori, data1, label='Data 1') # Data 1 
# plt.bar(kategori, data2, label='Data 2') # Data 2
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

<h3> Grouped Bar Plot

```python
data1 = [40, 10, 75, 35, 100]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']

x = np.arange(len(kategori))
x
```

```python
width = 0.4

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori)

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Grouped Bar Plot')

plt.legend()
plt.show()
```

<h1> Pie Plot / Pie Chart

```python
# Import modules
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

<h3> Simple Pie Plot

```python
kategori = ['A','B','C','D','E']
data = [215, 130, 245, 210, 100]
```

```python
plt.pie(data,
        labels=kategori,   # label
        autopct='%1.1f%%', # persentase
        startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

<h3>Pengaturan Legend dan Color

```python
kategori = ['A','B','C','D','E']
data = [215, 130, 245, 210, 100]
warna = ['pink','cyan','#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',
        startangle=45)

plt.legend() #Tampilan legend
plt.title('Contoh Pie Plot')

plt.show()
```

<h3> Exploded Pie Plot

```python
kategori = ['A','B','C','D','E']
data = [215, 130, 245, 210, 100]
warna = ['gray','cyan','#e03364', 'aquamarine', 'skyblue']
explode_var = [0.,0.,0.2,0.,0.08]
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=45)

plt.legend() #Tampilan legend
plt.title('Contoh Pie Plot')

plt.show()
```

<h1> Table Plot / Table Chart

```python
# Import Modules
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

<h3>Sample Data

```python
data_uas=[
    ['Deni', 70],
    ['Beni', 80],
    ['Serly', 90],
    ['Cecep', 75],
    ['Tio', 64],
     
]
```

```python
data_uas
```

<h3> Table Plot dengan OO Style

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                  loc='center')

table.set_fontsize(14)
table.scale(0.5,5) # ukuran kolom , baris

ax.axis(False) 

plt.show()
```

<h3>Table Plot dengan pyplot Style

```python
table = plt.table(cellText=data_uas,
                  loc='center')

table.set_fontsize(14)
table.scale(0.5,5) # ukuran kolom, baris

ax = plt.gca()
ax.axis(False)

plt.show()
```
