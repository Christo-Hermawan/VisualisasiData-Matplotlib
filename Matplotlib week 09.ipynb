<h3>Predefined Styles

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1,10,15)

plt.plot(x,x, 'bs-', label='lin')
plt.plot(x,np.log(x), 'go-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

<h3> Menerapkan multiple style

```python
plt.style.use(['seaborn', 'dark_background'])
```

```python
x = np.linspace(1,10,15)

plt.plot(x,x, 'bs-', label='lin')
plt.plot(x,np.log(x), 'go-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

<h3> Menerapkan temporary styling

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0,2 * np.pi)), 'b--')
plt.show()
```

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0,2 * np.pi)), 'r--')
    
plt.show()
```
