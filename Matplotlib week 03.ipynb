<h3> Simple Line Plot

```python
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
s = np.sin(2 * np.pi * x)
s
```

```python
fig, ax = plt.subplots() # OO style 

ax.plot(x, s)
ax.set(xlabel='nilai x',
       ylabel='nilai sin(x)',
       title='Visualisasi nilai sin')
ax.grid()

plt.show()
```

<h3> Multiple Subplots

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
x1
```

<h3> Multiple Subplots dengan OO Style

```python
fig, (ax1, ax2) = plt.subplots(2,1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'bo-') 
ax1.set_ylabel('Nilai cos(x1)')

ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x_2)')

ax2.set_xlabel('Nilai x')

plt.show()
```

<h3> Multiple Subplots dengan pyplot Style

```python
plt.subplot(2,1,1)
plt.plot(x1,y1, 'r*-')
plt.title('Multiple Subplots')
plt.ylabel('Nilai $cos(x_1)$')

plt.subplot(2,1,2)
plt.plot(x2,y2,'b.-')
plt.ylabel('Nilai $cos(x_2)$')
plt.xlabel('Nilai$x$')

plt.show()
```

<h1> Histogram

```python
# Import modules

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

<h3> Sample Data

```python
mu , sigma = 100, 15 # nilai mean & nilai standard deviation

x = mu + sigma * np.random.randn(10000) # normal distribution
x
```

```python
x.shape
```

<h3>Histogram dengan pyplot Style

```python
plt.hist(x,
         bins=50,
         facecolor='b',
         alpha=0.75) # 0.75 -> 75%

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, '$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

<h3> Histogram dengan OO Style

```python
fig, ax = plt.subplots()

ax.hist(x,
        bins=50,
        facecolor='r',
        alpha=0.50)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, '$\mu=100,\ \sigma=15$')
ax.grid() # Memunculkan grid

plt.show()
```
