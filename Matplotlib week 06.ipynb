

<h1> Box Plot / Box Chart

```python
#Import Modules
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

<h3>Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

<h3> Simple Box Plot

```python
plt.boxplot(data, labels=['Data'])

plt.title('Simple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.grid()
plt.show()
```

<h3>Notched Box Plot

```python
plt.boxplot(data, labels=['Data'], notch=True)

plt.title('Notched Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

<h3> Pengaturan marker untuk outlier

```python
plt.boxplot(data,
            labels=['Data'],
            showfliers=True,
            flierprops={'markerfacecolor':'b','marker':'s'})

plt.title('Pengaturan Marker')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

<h3> Horizontal Box Plot

```python
plt.boxplot(data, labels=['Data'], vert=False) #vert=True menampilkan jadi vertikal

plt.title('Horizontal Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

<h3> Multiple Box Plot

```python
np.random.seed(10)

data1 = np.random.normal(70, 25, 200)
data2 = np.random.normal(80, 30, 200)
data3 = np.random.normal(100, 10, 200)
data4 = np.random.normal(90, 20, 200)


data = [data1, data2, data3, data4]
labels = ['Data 1', 'Data 2', 'Data 3', 'Data 4']
```

```python
plt.boxplot(data, labels=labels)

plt.title('Multiple Box Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

<h1> Violin Plot / Violin Chart

```python
#Import Modules
```

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

<h3> Sample Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

<h3> Simple Violin Plot

```python
plt.violinplot(data)

plt.title('Sample Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

<h3> Pengaturan pada violin plot

```python
plt.violinplot(data,
               showextrema=True,
               showmeans=False,
               showmedians=False,
               quantiles=[0.25, 0.5, 0.75])

plt.title('Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

<h3> Horizontal Violin Plot

```python
plt.violinplot(data, vert=False)

plt.title('Horizontal Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

<h3> Multiple Violin Plot

```python
np.random.seed(10)

data1 = np.random.normal(90,20,200)
data2 = np.random.normal(80,30,200)
data3 = np.random.normal(70,25,200)
data4 = np.random.normal(100,10,200)

data = [data1,data2,data3,data4]
```

```python
plt.violinplot(data)

plt.title('Multiple Violin Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

<h1> Twin Axes

```python
#Import Modules
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

<h3> Kasus Twin Axes 1

```python
x = np.linspace(1, 10, 25)
x
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), 'y*-', label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x), 'go-', label='log')
ax2.set_ylabel('log')

fig.suptitle('Contoh Twin Axes')
fig.legend(loc='upper left')

plt.show()
```

```python
x = np.random.randint(16 ,40, size=30)
x
```

```python
def C2F(celsius=0):
    return(celsius * 1.8) + 32

def konversi_sumbu(ax1):
    y1, y2 = ax1.get_ylim()
    ax2.set_ylim(C2F(y1), C2F(y2))
    ax2.figure.canvas.draw()
    
fig, ax1 = plt.subplots()
ax2 =ax1.twinx()

ax1.callbacks.connect('ylim_changed', konversi_sumbu)
ax1.plot(x)

ax1.set_xlabel('Hari')
ax1.set_ylabel('Celsius')
ax2.set_ylabel('Fahrenheit')

fig.suptitle('Temperatur Udara')
plt.show()
```
