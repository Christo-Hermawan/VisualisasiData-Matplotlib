<h3> Simple Scatter Plot

```python
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
```

```python
plt.scatter(x, y, label='Data 1', color='b')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Simple Scatter Plot')

plt.legend()

plt.show()
```

<h3> Multiple Scatter Plot

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [5,7.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,6]
```

```python
plt.scatter(x1, y1, label='Data 1', color='b')
plt.scatter(x2, y2, label='Data 2', color='g')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Multiple Scatter Plot')

plt.legend()

plt.show()
```

<h3> Pengaturan Marker

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [5,7.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,6]
```

```python
plt.scatter(x1,y1,
            color='cyan',
            linewidths=1,
            marker='^',
            edgecolor='black',
            s=100)
plt.scatter(x2,y2,
            color='red',
            linewidths=1,
            marker='*',
            edgecolor='purple',
            s=200)
```

<h3> Scatter Plot untuk data multi dimensi

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris['data'].T
```

```python
iris.keys()
```

```python
iris['data']
```

```python
iris['feature_names']
```

```python
features = iris['data'].T
features
```

```python
iris['target']
```

```python
iris['target_names']
```

```python
plt.scatter(features[0], features[1],
            alpha=0.2,
            s=100*features[3],
            c=iris['target'],
            cmap='viridis')

plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
plt.title('Iris Dataset')

plt.colorbar()

plt.show()
```

<h3> Fungsi Plotting

```python
def gambar_barplot():
    data1 = [25,85,75,40,60]
    data2 = [40,35,20,55,10]
    kategori = ['A','B','C','D','E']
    
    x = np.arange(len(kategori))
    
    width = 0.35
    plt.bar(x-width/2, data1, width, label='Data 1')
    plt.bar(x+width/2, data2, width, label='Data 2')
    
    plt.xticks(x, kategori)
    
    plt.grid(linestyle='--',
             linewidth=1,
             axis='y',
             alpha=0.75)
    
    plt.xlabel('Kategori')
    plt.ylabel('Jumlah')
    plt.title('Contoh Grouped Bar Plot')
    
    plt.legend()
    plt.show()
```

```python
 gambar_barplot()
```
