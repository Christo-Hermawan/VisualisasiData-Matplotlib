```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

<h3> Sample dataset

```python
data = {'Item A': 109438,
        'Item B': 112512,
        'Item C': 103219,
        'Item D': 123122,
        'Item E': 139958,
        'Item F': 102949,
        'Item G': 124856,
        'Item H': 112479,
        'Item I': 139374,
        'Item J': 124856}
data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

<h3> Simple Plot

```python
fig, ax = plt.subplots()

ax.barh(items, count)

plt.show()
```

<h3> Pengaturan Style

```python
plt.style.available
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

<h3> Pengaturan Tick Label

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

# set properties
plt.setp(labels,
         rotation=45,
         horizontalalignment='right')

plt.show()
```

<h3> Pengaturan Format pada Ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

<h3>Pengaturan Label pada Sumbu (axis) dan Judul (title)

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')
plt.show()
```

<h3> Penambahan Garis (vertical/horizontal line) pada Plot

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
           ls='--',
           color='b')

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

<h3> Menyimpan Hasil Plot ke dalam suatu File

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
           ls='--',
           color='b')

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

fig.savefig('sales.png',
            transparent=False,
            dpi=80)
```
